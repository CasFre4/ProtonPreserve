```python
import requests
from bs4 import BeautifulSoup
import numbers
import os
```

```python
def isNumber(obj):
    return isinstance(obj, numbers.Number) and not isinstance(obj, bool)
# URL that contains the table
baseurl = "https://repo.protonvpn.com/debian/"

# Fetch the page
response = requests.get(baseurl)
response.raise_for_status()

# Parse the HTML
soup = BeautifulSoup(response.text, "html.parser")
# print(soup)
# Find the first table
# soup.__iter__

search = [baseurl]
file_sizes = []
to_download = []
while search:
    url = search.pop()
    # Fetch the page
    response = requests.get(url)
    response.raise_for_status()

    # Parse the HTML
    soup = BeautifulSoup(response.text, "html.parser")
    for a in soup.find_all('a'):
        bb_bro = a.next_sibling
        href = a.get('href')
        if href=='../':
            continue
        # print(href[-1:])
        if href[-1:]=="/":
            isdir = True
        else:
            isdir=False
        if bb_bro.strip()[-1:].isnumeric():
            num_end = True
        else:
            num_end = False
        if (num_end and isdir):
            raise Exception(f'Ends in number: {num_end} and is dir: {isdir}\n{a} {bb_bro}')
        if not num_end and not isdir:
            raise Exception(f'No end number: {num_end} and is file: {isdir}\n{a} {bb_bro}')
        if isdir:
            search.append(url + href)
        elif num_end:
            to_download.append(url + href)
            digit_len = 1
            end_number_string = bb_bro.strip()[-1 * digit_len:]
            while bb_bro.strip()[-1*digit_len:].isnumeric():
                digit_len += 1
                end_number_string = bb_bro.strip()[-1*digit_len:]
            end_number = int(end_number_string)
            file_sizes.append(end_number)
        else:
            raise Exception(f'Error: Unexpected result {a} {bb_bro}')
        # print(a)
        print(f'href: {href}, isdir: {isdir}, num_end: {num_end}\n')
        # print(a.next_sibling.strip()[-1:].isnumeric())
        
        # print(isNumber(a.next_sibling.strip()[-1:]))
        # print([item for item in a.next_siblings])
        # print(type(line))
        # print(bool(line.find('a')))
        # if line.find('a'):
        #     print(line)
print(f'GB: {sum(file_sizes)/1000000000} KB: {sum(file_sizes)/1000}, Total File Sizes: {sum(file_sizes)}')
```

```python
pref = 'https://repo.protonvpn.com/'
count = 0
for url in to_download:
    if url.startswith(pref):
        path = 'files/' + url.removeprefix(pref)
        dir_path = os.path.dirname(path)
        os.makedirs(dir_path, exist_ok=True)
        response = requests.get(url)
        try:
            with open(path, "wb") as f:
                f.write(response.content)
        except Exception as e:
            raise Exception(f"Broker: {e}")
            
    else:
        raise Exception(f"Doesnt start with {pref}\nURL: {url}")
    count += 1
    print(f'Finished File {count}/{len(to_download) + 1}')
```
